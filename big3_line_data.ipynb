```python
rankings_00 = pd.read_csv('atp_rankings_00s.csv')
rankings_10 = pd.read_csv('atp_rankings_10s.csv')
rankings_curr = pd.read_csv('atp_rankings_current.csv')
players = pd.read_csv('atp_players.csv')
```

```python
players.columns = ['player_id', 'first_name', 'last_name', 'hand', 'birth_date', 'country_code']
rankings_curr.columns = ['ranking_date', 'rank', 'player', 'points']
rankings_curr = rankings_curr.drop(range(0,678))
```

```python
three_dict = {104745: 'Rafael Nadal', 103819: 'Roger Federer', 104925: 'Novak Djokovic'}
```

```python
rankings = rankings_00.append([rankings_10, rankings_curr]).reset_index()
rankings = rankings.drop(columns=['index'])
```

```python
big_three_rank = rankings.loc[rankings['player'].isin(three_dict.keys())].reset_index()
```

```python
big_three_rank['player'] = big_three_rank['player'].map(three_dict)
```

```python
dates = [20041206, 20050606, 20051205, 20060605, 20061204, 20070604, 20071203, 20080602, 20081201, 20090601, 20091207,
        20100607, 20101206, 20110606, 20111205, 20120611, 20121203, 20130610, 20131202, 20140609, 20141201,
        20150608, 20151207, 20160606, 20161205, 20170612, 20171204, 20180611, 20181203, 20190610, 20191202]
```

```python
big_three_rank.loc[(big_three_rank['ranking_date'] >= 20120601) & (big_three_rank['ranking_date'] <= 20121210)]
```

```python
fed_rank = big_three_rank.loc[big_three_rank['player'] == 'Roger Federer']
fed_rank.loc[fed_rank['ranking_date'].isin(dates)]
```

```python
rank_date = []
for d in dates:
    rank_date.append(dp.parse(str(d)))
```

```python
big_three_rank['ranking_date'] = rank_date
```

```python
big_three_rank = big_three_rank.reset_index()
```
